```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from urllib.request import urlopen
from bs4 import BeautifulSoup
```

Get the html of the page

```python
url = "http://www.hubertiming.com/results/2017GPTR10K"
html = urlopen(url)
```

```python
soup = BeautifulSoup(html, 'lxml')
type(soup)
```

```python
soup
```

```python
text = soup.get_text()
text
```

```python
rows = soup.find_all('tr')
```

4th row is the header

```python
rows[4]
```

get a list with all the rows after the header, without the html tags

```python
rows_list = []
for row in rows[5:]:
    fields = row.find_all('td')
    # clean up the text of html tags
    rows_list.append(BeautifulSoup(str(fields), 'lxml').get_text())
rows_list[:5]
```

create a dataframe with all these rows

```python
df = pd.DataFrame(rows_list)
df.head(10)
```

remove the weird characters and split the fields

```python
df1 = df[0].str.strip('[').str.strip(']').str.split(',', expand=True)
df1.head(10)
```

get the labels (from the 'th' tag, could also be by the 4th row)

```python
col_labels = soup.find_all('th')
col_labels = BeautifulSoup(str(col_labels), 'lxml').get_text()
col_labels
```

use the labels as headers for the dataframe

```python
df1.columns = col_labels.strip('[').strip(']').split(',')
```
